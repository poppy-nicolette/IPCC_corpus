```python
import csv
def readCSV(dir, fileName, delim):
    with open(dir + fileName) as csvFile:
        isHeader = True
        reader = csv.reader(csvFile, delimiter = delim)
        result = []
        for row in reader:
            if(isHeader):
                isHeader = False
            else:
                result.append(row)
        return result
```

```python
import urllib.error
import urllib.request
import json
def requestAPI(entry):
    requestString = "https://api.openalex.org/works/" + entry + "?select=id,keywords,sustainable_development_goals"
    try:
        with urllib.request.urlopen(requestString) as netStream:
            return json.load(netStream)

    except urllib.error.URLError as e:
        print("Error", e)
```

```python
def partition(originalArray):
    #https://favtutor.com/blogs/partition-list-python

    n = 100
    
    # partition the list using list comprehension
    partitionArray = [originalArray[i:i + n] for i in range(0, len(originalArray), n)]
    return partitionArray
```

```python
def writeJSON(dir, fileName, content):
    with open(dir + fileName, "w") as file:
        file.write(content)
```

```python
ids = list(map(lambda x: x[0].split("/")[3], readCSV("./tables/tables/", "core_works.csv", ",")))
print(len(ids))
partitionArray = partition(ids)
```

```python
counter = 0
intermedateName = "./intermediateData/"
for items in partitionArray:
    fileName = "intermediateData_" + str(counter* 100) +"_" + str((counter + 1) * 100) +".json"
    counter += 1
    print("Calling API for ", fileName, "...")
    result = list(map(requestAPI, items))
    print("Writing file for",  fileName, "...")
    writeJSON(intermedateName, fileName, json.dumps(result))
```

```python
def readJSON(dir, fileName):
    with open(dir + fileName, "r") as file:
        return json.load(file)
```
