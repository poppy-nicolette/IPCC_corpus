```python
# import core works
import pandas as pd
import os
import plotly.express as px
```

```python
df = pd.read_csv("/Users/poppyriddle/Downloads/tables/works.csv")
df.drop(columns=['abstract'], inplace=True)
df.head(2)

# import core_corks.csv
df_core = pd.read_csv("/Users/poppyriddle/Downloads/tables/core_works.csv")
#df_core.head(2)
# merge inner on core_works.csv
df_merged_core = df_core.merge(df,on='openalex_work_id', how='inner')
print(len(df_merged_core))
df_merged_core.head(2)

```

```python
pub_year_df = pd.DataFrame(df_merged_core.groupby(df_merged_core['publication_year'])['publication_year'].count()).unstack()

pub_year_df = pub_year_df.reset_index()
pub_year_df.drop(columns=['level_0'], inplace=True)
pub_year_df = pub_year_df.rename(columns={'publication_year':'year',0:'count'})
pub_year_df
```

```python

fig = px.bar(pub_year_df,
            x="year",
            y='count',
            title="Core Works by year, 1995-2020",
            template="simple_white",
            labels={ # replaces default labels by column name
                "count": "Count",  "year": "Published year"},)
fig.show()
```

```python
# doc types on df_merged_core
doc_type_df = pd.DataFrame(df_merged_core.groupby(df_merged_core['type'])['type'].count()).unstack()
doc_type_df = doc_type_df.reset_index()
doc_type_df.drop(columns=['level_0'], inplace=True)
doc_type_df = doc_type_df.rename(columns={'type':'type',0:'count'})
doc_type_df = doc_type_df.sort_values('count', ascending=False)
doc_type_df.to_excel("tables/doc_type.xlsx")
doc_type_df
```

```python
fig = px.bar(doc_type_df,
            x="type",
            y='count',
            title="Count of Document types",
            text_auto=True,
            template="simple_white",
            labels={ # replaces default labels by column name
                "count": "Count",  "type": "Document type"},
            category_orders={ # replaces default order by column name
                "type": ["article","review","book-chapter","book","report","letter","preprint","other","editorial","erratum","dissertation","reference-entry","paratext"]},)
fig.update_traces(textfont_size=12, textangle=0, textposition="outside", cliponaxis=False)
fig.show()
```

```python
# journals source_id	source_name
source_df = pd.DataFrame(df_merged_core.groupby(['source_name'])['source_id'].count()).unstack()

source_df = source_df.reset_index()
source_df.drop(columns=['level_0'], inplace=True)
source_df = source_df.rename(columns={'source_name':'source_name',0:'count'})
source_df = source_df.sort_values('count', ascending=False)
#source_df = source_df.head(10)
#source_df.to_excel("tables/source_names.xlsx")
source_df
```

```python
# new df /Users/poppyriddle/Documents/Github/IPCC_corpus/Corpus/descriptive_analysis/data/core_oa_data.csv
core_oa_df = pd.read_csv("/Users/poppyriddle/Documents/Github/IPCC_corpus/Corpus/descriptive_analysis/data/core_oa_data.csv", encoding='UTF-8')
core_oa_df.head(5)
```

```python
# oa status
oa_status_df = pd.DataFrame(core_oa_df.groupby(['oa_status'])['oa_status'].count()).unstack()
oa_status_df = oa_status_df.reset_index()
oa_status_df.drop(columns=['level_0'], inplace=True)
oa_status_df = oa_status_df.rename(columns={'oa_status':'oa_status',0:'Count'})
oa_status_df = oa_status_df.sort_values('Count', ascending=False)

# add percentages column
oa_status_df['Percentage'] = oa_status_df['Count'].apply(lambda x: x/(oa_status_df['Count'].sum())*100)

oa_status_df.to_excel("tables/oa_status.xlsx")
oa_status_df
```

```python
# countries and affiliations
country_df = pd.read_csv("/Users/poppyriddle/Documents/Github/IPCC_corpus/Corpus/descriptive_analysis/data/core_affiliation_data.csv", encoding="UTF-8")
country_df.head(5)
```

```python
country_code_df = pd.DataFrame(country_df.groupby(['institution_country_code'])['institution_country_code'].count()).unstack()
country_code_df = country_code_df.reset_index()
country_code_df.drop(columns=["level_0"],inplace=True)
country_code_df = country_code_df.rename(columns={'institution_country_code':'country_code',0:'Count'})
country_code_df.to_excel("tables/country_code.xlsx")
country_code_df
```

```python
# Plotly chloropleth
# load ISO-3166 codes
iso_codes_df = pd.read_excel("/Users/poppyriddle/Documents/Github/IPCC_corpus/Corpus/descriptive_analysis/ISO_3166_country_codes.xlsx")
iso_codes_df = iso_codes_df[['A2_code','A3_code','ISO 3166']]
iso_codes_df

country_code_df_merged = country_code_df.merge(iso_codes_df, left_on='country_code', right_on='A2_code', how='inner')
country_code_df_merged.drop(columns=['A2_code'],inplace=True)
country_code_df_merged
```

```python
country_code_merged_export = country_code_df_merged.sort_values('Count', ascending=False)
country_code_merged_export.to_excel("tables/country_codes_export.xlsx")
country_code_merged_export
```

```python
#plotly chloropleth with country_code_df_merged
import plotly.express as px

fig = px.choropleth(country_code_df_merged,
                    locations="A3_code",
                    color="Count", 
                    hover_name="ISO 3166",
                    locationmode='ISO-3',
                    title="Authors on works by country code associated with institutional affiliation",
                    color_continuous_scale=px.colors.sequential.Plotly3)
fig.show()
```

```python
# top affiliations 
institution_df = pd.read_csv("/Users/poppyriddle/Documents/Github/IPCC_corpus/Corpus/descriptive_analysis/data/core_affiliation_data.csv", encoding="UTF-8")
institution_df.head(5)
top_inst_df = pd.DataFrame(institution_df.groupby(['institution_display_name'])['institution_display_name'].count()).unstack()
top_inst_df = top_inst_df.reset_index()
top_inst_df.drop(columns=['level_0'], inplace=True)
top_inst_df = top_inst_df.rename(columns={'institution_display_name':'Affiliation',0:'Count'})
top_inst_df = top_inst_df.sort_values('Count', ascending=False)
top_inst_df.head(10).to_excel("tables/top_inst.xlsx")
top_inst_df
```
