```plaintext
# open data/core_works.csv as df_core_works
df_core_works = pd.read_csv('data/core_works.csv')

#look up all files in folder that start with 'nodes_*.csv'
#list files in the directory and filter those that start with "nodes_"
net_files = [f for f in os.listdir("networks/") if f.startswith("nodes_")]
for file in net_files:
    #read the CSV file into a DataFrame
    df = pd.read_csv(os.path.join("networks", file))
    #for each in list, perform above function)
    # create new column in df_nodes_bc called 'core_work'
    df['core_work'] = 0

    # join df_nodes_bc with df_core_works left_on='Label', right_on='openalex_work_id', how='left'
    df = df.merge(df_core_works, left_on='Label', right_on='openalex_work_id', how='left')

    #not_null returns True/false and astype() returns 0 for False, 1 for True
    df.loc[df['openalex_work_id'].notnull(), 'core_work'] = 1
    #drop unnecessary cols
    df.drop(columns=['openalex_work_id'], inplace=True)

    #save out each to networks overwriting the original
    output_file = os.path.join("networks2", file)
    df.to_csv(output_file, index=False)
    print(Fore.LIGHTGREEN_EX + f"✅ Exported file as {output_file}" + Style.RESET_ALL)
```
