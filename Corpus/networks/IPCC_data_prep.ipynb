#IPCC corpus project
- [ ] open citations.csv and convert openalex_ids to id from works.csv


```plaintext
import pandas as pd
```

```plaintext
citations_df = pd.read_csv("data/citations.csv",encoding='utf-8')
works_df = pd.read_csv("data/works.csv", encoding='utf-8')
```

```plaintext
#create id column of sequential numbers
works_df.insert(1,'id',range(1,len(works_df)+1))
```

```plaintext
#rename citations_df column names
works_df = works_df.rename(columns={'openalex_work_id':'openalex_id','id':'id'})
#merge on openalex_id
citations_df = citations_df.merge(works_df, left_on='citing_id', right_on='openalex_id', how='inner')
```

```plaintext

#clean up unneeded columns and rename - probably a better way to do this
citations_df.drop(columns=['citing_id','openalex_id'], inplace=True)
citations_df = citations_df.rename(columns={'cited_id':'cited_id','id':'citing_id'})
```

```plaintext
citations_df = citations_df.merge(works_df, left_on='cited_id', right_on='openalex_id', how='inner')
```

```plaintext
citations_df.drop(columns=['cited_id', 'openalex_id'], inplace=True)
citations_df = citations_df.rename(columns={'citing_id':'citing_id', 'id':'cited_id'})
```

```plaintext
citations_df.to_csv('data/citations.csv',encoding='utf-8',index=False)
works_df.to_csv('data/works.csv',encoding='utf-8',index=False)
```
