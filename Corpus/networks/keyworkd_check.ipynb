# add kw from OpenAlex to nodes_*.csv files
This adds keywords extracted from OpenAlex to the nodes_*.csv files created by biblionet.py


```python
import pandas as pd
import os
```

```python
df = pd.read_json("/Users/poppyriddle/Documents/Github/IPCC_corpus/Corpus/labels/keywords_SDGs.json")
sdg_df = df[['sustainable_development_goals','keywords']]
sdg_df.iloc[0,0]
df
```

```python
df2 = df['keywords'].apply(pd.Series)
df2
```

```python
df2.fillna(inplace=True, value=0)

df3 = df2.map(lambda x: list(x.values()) if x!=0 else 0)
df4 = df3.map(lambda x: x[1] if x != 0 and len(x)>1 else None)
df4.head(20)
df4.to_excel("keywords2.xlsx")
df4.head(20)
```
