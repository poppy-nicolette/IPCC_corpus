```python
df = pd.read_json("/Users/poppyriddle/Documents/Github/IPCC_corpus/Corpus/labels/keywords_SDGs.json")
sdg_df = df[['sustainable_development_goals']]
sdg_df.iloc[0,0]
sdg_df.fillna(inplace=True, value=0)
sdg_df2 = sdg_df.map(lambda x: x[0] if len(x)>0 else None)
sdg_df3 = sdg_df2.map(lambda x: x.get('display_name') if x!=None else None)
sdg_df_merged = pd.concat([id_df,sdg_df3], axis=1)
#sdg_df_merged.fillna(inplace=True, value='None')
sdg_df_merged = sdg_df_merged.rename(columns={'id':'oa_id', 'sustainable_development_goals':'sdgs'})
sdg_df_merged.to_excel("sdgs.xlsx",index=False)
sdg_df_merged
```

```python
#look up all files in folder that start with 'nodes_*.csv'
#list files in the directory and filter those that start with "nodes_"
# merge with sdg_df_merged on 'id'
net_files = [f for f in os.listdir("networks_copy/") if f.startswith("nodes_")]
for file in net_files:
    #read the CSV file into a DataFrame
    df = pd.read_csv(os.path.join("networks_copy", file))
    #for each in list, perform above function)

    # join df_nodes_bc with df_core_works left_on='Label', right_on='openalex_work_id', how='left'
    df = df.merge(sdg_df_merged, left_on='Label', right_on='oa_id', how='left')


    #save out each to networks overwriting the original
    output_file = os.path.join("networks_copy", file)
    df.to_csv(output_file, index=False)
    print(Fore.LIGHTGREEN_EX + f"✅ Exported file as {output_file}" + Style.RESET_ALL)
```
